### Open challenger datasets

> Insert here the code that opens the challenger dataset as `challenger_dataset: xarray.Dataset`

```python
from datetime import datetime

import xarray
from oceanbench.core.challenger_datasets import _open_multizarr_forecasts_as_challenger_dataset


def _glonet2_correct_stats_dataset_path(start_datetime: datetime) -> str:
    start_datetime_string = start_datetime.strftime("%Y%m%d")
    return (
        "https://s3.waw3-1.cloudferro.com/oceanbench-bucket/"
        f"dev/ml-forecast-outputs-glonet2-correct-stats/glonet2/{start_datetime_string}.zarr"
    )


challenger_dataset: xarray.Dataset = _open_multizarr_forecasts_as_challenger_dataset(
    _glonet2_correct_stats_dataset_path
)
challenger_dataset
```

### Evaluation configuration

```python
region = 'global'
```

### Evaluation of challenger dataset using OceanBench

#### Root Mean Square Deviation (RMSD) of variables compared to GLORYS reanalysis

```python
oceanbench.metrics.rmsd_of_variables_compared_to_glorys_reanalysis(
    challenger_dataset,
    region=region,
)
```
